```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./vinos/iris.csv")
data
```

# Arboles de clasificacion

```python
data["Species"].unique()
```

```python
data.shape
```

```python
plt.hist(data.Species) #ver como estan distribuidas
```

```python
data["is_train"] = np.random.uniform(0,1,len(data))<= 0.75
data.head()
```

```python
train = data[data["is_train"]==True]
```

```python
train
```

```python
test = data[data["is_train"]==False]
test
```

```python
test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion="entropy", min_samples_split=20, random_state=99)
```

```python
tree
```

```python
predictors = ["Sepal.Length", "Sepal.Width","Petal.Length","Petal.Width"]
```

```python
preds = tree.fit(train[predictors], train["Species"])
preds
```

```python
preds= tree.predict(test[predictors])
```

```python
preds
```

```python
pd.crosstab(test["Species"],preds,rownames=["Obervado"], colnames=["Predicciones"])
```
