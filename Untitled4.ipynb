```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size=[100,])
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]], size=[50,])
X = np.concatenate((a,b))
plt.scatter(X[:,0],X[:,1])
```

```python
Z = linkage(X, "ward")
fig = plt.figure(figsize=(25,10))
dendrogram(Z, leaf_font_size=8.0, leaf_rotation=90.0, truncate_mode="lastp", p=12, show_leaf_counts=True, show_contracted=True)
plt.show()
```

```python
last_zs = Z[-12:,2]
last_zs_rev = last_zs[::-1]
last_zs_rev
```

```python
nro_clusts = np.arange(1, len(last_zs_rev)+1)
nro_clusts
```

```python
plt.plot(nro_clusts, last_zs_rev)
```
